```python
import matplotlib.pyplot as plt
```

```python
x = ['Jan' , 'Feb' , 'Mar' , 'Apr' , 'May' , 'Jun' ]
inflation = [5, 5.1, 5.3, 5.7, 6, 7]
```

```python
plt.plot(x, inflation)
```

```python
plt.plot(x, inflation, 'o-', color = 'red')
```

```python
plt.plot(x, inflation, 'o-', color = 'red')
plt.xlabel("Month")
plt.ylabel("Inflation")
plt.show()  # teha lekheko gayaeb hunxa
```

# bar

```python
party = ['RPP' , 'RSP' , 'Maobadi', 'Congress' , 'CPN-UML']
seat = [14,24,32,89,79]
```

```python
plt.bar(x=party, height=seat, color = ['red' , 'green' , 'purple' , 'orange' ,'black'])
```

```python
plt.pie(seat , labels=party)
plt.show
```

```python
plt.pie(seat , labels=party, autopct ='%.2f%%', colors = ['blue' , 'cyan' , 'gray', 'yellow', 'brown'])
plt.show()
```

```python
import seaborn as sns
sns.set_theme()
tips = sns.load_dataset("tips")
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
plt.show()
```
